```python
import yfinance as yf
import pandas as pd

# Adjustment made to section the date back to the creation of the DRIV fund so we dont have issues later when cleaning
start_date = '2018-04-20'
end_date = '2024-12-20'

tsla = yf.Ticker("TSLA")
tsla_data = tsla.history(start=start_date, end= end_date)

ndx = yf.Ticker("^NDX")
ndx_data = ndx.history(start=start_date, end= end_date)

driv = yf.Ticker("DRIV")
driv_data = driv.history(start=start_date, end= end_date)

xly = yf.Ticker("XLY")
xly_data = xly.history(start=start_date, end= end_date)

# Change made to accomodate LIT into our composite index 
# LIT is the global lithium & battery ETF, very useful to have
lit = yf.Ticker("LIT")
lit_data = lit.history(start=start_date, end= end_date)

print(tsla_data) 
print(ndx_data)
print(driv_data)
print(xly_data)
print(lit_data)
```

```python
tsla_data.to_csv('TSLA_data1.csv')
ndx_data.to_csv('NDX_data1.csv')
driv_data.to_csv('DRIV_data1.csv')
xly_data.to_csv('XLY_data1.csv')
```

```python
lit_data.to_csv('LIT_data1.csv')
```
